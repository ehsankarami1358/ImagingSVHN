```python
from transformers import*
```

```python
src='en'
dst='de'
task_name=f'translation_{src}_to_{dst}'
model_name=f'Helsinki-NLP/opus-mt-{src}-{dst}'
translator=pipeline(task_name,model=model_name,tokenizer=model_name)
```

```python
translator('You are a genius.')[0]['translation_text']
```

```python
def get_translation_model_and_tokeneizer(src_lang,dst_lang):
  model_name=f'Helsinki-NLP/opus-mt-{src}-{dst}'
  tokenizer=AutoTokenizer.from_pretrained(model_name)
  model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
  return model,tokenizer
```

```python
src='en'
dst='zh'

model,tokenizer=get_translation_model_and_tokeneizer(src,dst)
```

```python
inputs=tokenizer.encode('You are a genius.',return_tensors='pt',max_length=512,truncation=True)
print(inputs)
```

```python
greedy_outputs=model.generate(inputs,max_length=40,num_beams=4,early_stopping=True)
print(tokenizer.decode(greedy_outputs[0],skip_special_tokens=True))
```

```python
beam_outputs=model.generate(inputs,max_length=40,num_beams=4,early_stopping=True,num_return_sequences=3)
print(tokenizer.decode(beam_outputs[0],skip_special_tokens=True))
```
